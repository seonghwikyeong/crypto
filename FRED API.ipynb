```python
import requests
import pandas as pd
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Your FRED API key
API_KEY = "0f38e32c2e981eed28544328383782cf"

# CPI data series ID (e.g., CPI for All Urban Consumers: All Items in U.S. City Average)
SERIES_ID = "PAYEMS"

# API endpoint
url = f"https://api.stlouisfed.org/fred/series/observations"

# Parameters for the API request
params = {
    "series_id": SERIES_ID,
    "api_key": API_KEY,
    "file_type": "json",
    "frequency": "m",  # Monthly data
    "observation_start": "2020-01-01",  # Adjust start date as needed
}

# Fetch data from FRED API
response = requests.get(url, params=params, verify=False)

if response.status_code == 200:
    data = response.json()
    # Parse observations into a DataFrame
    observations = data["observations"]
    df = pd.DataFrame(observations)
    df["date"] = pd.to_datetime(df["date"])
    df["value"] = pd.to_numeric(df["value"])
    print(df.head())
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
    print(response.text)
```
